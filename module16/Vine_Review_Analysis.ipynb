```python
import pandas as pd
```

```python
table_df = pd.read_csv("vine_table.csv")
```

```python
table_df
```

```python
table_df.dropna()
```

```python
review_table = table_df["total_votes"] >= 20
review_table
```

```python
review_table_df = table_df[review_table]
review_table_df
```

```python
helpful_votes_percentage = (table_df["helpful_votes"]/table_df["total_votes"])*100
helpful_votes_percentage
```

```python
review_table_df.insert(1, "helpful_percentage", helpful_votes_percentage)
```

```python
filter_df = review_table_df["helpful_percentage"] >= 50
filter_df
```

```python
review_table_df = review_table_df[filter_df]
review_table_df
```

```python
vine_yes = review_table_df["vine"] == "Y"
```

```python
vine_no = review_table_df["vine"] == "N"
```

```python
vine_yes_df = review_table_df[vine_yes]
```

```python
vine_no_df = review_table_df[vine_no]
```

```python
vine_yes_df
```

```python
vine_no_df
```

```python
total_reviews = review_table_df["review_id"].count()
total_reviews
```

```python
total_5 = review_table_df.loc[(review_table_df["star_rating"] == 5)]
total_5
```

```python
# Total Vine reviews
total_yesvine_reviews = vine_yes_df["star_rating"].count()
total_yesvine_reviews
```

```python
# Total Non Vine Reviews
total_novine_reviews = vine_no_df["star_rating"].count()
total_novine_reviews
```

```python
yesfive_star_count = vine_yes_df.loc[(vine_yes_df["star_rating"] == 5), "star_rating"]
yesfive_star_count
```

```python
nofive_star_count = vine_no_df.loc[(vine_no_df["star_rating"] == 5), "star_rating"]
nofive_star_count
```

```python
# Non Vine reviews that are 5 stars
fivestarnovine_reviews = nofive_star_count.count()
fivestarnovine_reviews
```

```python
# Vine reviews that are 5 stars
fivestaryesvine_reviews = yesfive_star_count.count()
fivestaryesvine_reviews
```

```python
y_percentage = (fivestaryesvine_reviews/total_yesvine_reviews)*100
n_percentage = (fivestarnovine_reviews/total_novine_reviews)*100

print(f'Paid 5-Star Rating Percentage: {y_percentage} % \nUnpaid 5-Star Rating Percentage: {n_percentage} %')
```
